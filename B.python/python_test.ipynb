```python
def replace_words_with_digits(s):
    # 영단어와 숫자 매핑
    word_to_digit = {
        "zero": "0", "one": "1", "two": "2", "three": "3",
        "four": "4", "five": "5", "six": "6", "seven": "7",
        "eight": "8", "nine": "9"
    }

    # 변환 결과를 저장할 문자열
    result = ""

    # 문자열의 인덱스를 기준으로 처리
    i = 0
    while i < len(s):
        # 숫자일 경우 바로 결과에 추가
        if s[i].isdigit():
            result += s[i]
            i += 1
        else:
            # 영단어를 인식하기 위해 현재 위치부터 단어 끝까지 확인
            for word, digit in word_to_digit.items():
                if s[i:i+len(word)] == word:
                    result += digit
                    i += len(word)
                    break

    return result

# 예시 문자열
s = "77three4one"
# 원래 숫자를 찾기 위한 함수 호출
original_number = replace_words_with_digits(s)
print(original_number)

```

```python
# 7. 예제로 for문 익혀보기
#문제 : 절대값 정수가 담긴 정수 변수와 그 정수의 부호를 차례로 담은 부호 변수가 있다
# 두 변수를 이용해 실제 정수의 합을 구해주세요 (True : 양수, False : 음수)

정수=[10,5,8]
부호=[True,False,True]

for i in range(len(정수)):
    if 부호[i]:
        print(정수[i])
    else:
        print(-정수[i])

합=0

for k, v in zip(정수, 부호):
    if v:
        합+=k
    else:
        합-=k

print(f'총 합은 {합}입니다.')
```

```python
숫자 = "12344879498494"
# 전체 출력
print(숫자[::])
print(숫자[::-1])
# 첫번째 인덱스 기준으로 2칸씩 띄어쓰기 출력하기
print(숫자[::2])
# 첫번째 인덱스 기준으로 2칸씩 띄어쓰기 출력하기
print(숫자[4:1:-1])
print(숫자[1:4])
```

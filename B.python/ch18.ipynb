```python
class 클래스이름:
    def __init__(self, 속성값1, 속성값2):
        self.속성1 = 속성값1
        self.속성2 = 속성값2
```

```python
class Person:
    def __init__(self, 이름, 나이, 주소):
        self.hello = "안녕하세요"
        self.이름 = 이름
        self.나이 = 나이
        self.주소 = 주소

    def greeting(self):
        print('{0} 저는 {1}입니다.'.format(self.hello,self.이름))

maria = Person('마리아', 20, '서울시 서초구 서초동')
maria.greeting() # 안녕하세요 저는 마리아입니다.

print('이름:', maria.이름) # 마리아
print('나이:', maria.나이) #20
print('주소:', maria.주소) #서울시 서초구 반포동
```

```python
class Car(object):
    maxspeed = 300
    maxpeople = 5
    def move(self, x):
        print("스피드로 달리고 있습니다")
    def stop(self):
        print("멈췄습니다")

class HybridCar(Car):
    battery = 1000
    batteryKM = 300
class ElectricCar(HybridCar):
    battery = 1000
    batteryKM = 300

k5 = HybridCar()
k5.maxspeed

electricCark5 = ElectricCar()
electricCark5.maxspeed

#print(dir(k5))
```

###2. if문 문법과 주의사항(콜론, 들여쓰기)

````python
#2. if문 문법과 주의사항(콜론, 들여쓰기)
개수 = int(input("붕어빵 몇개 사실거에요?"))

if 개수 >3:
    print("개당 1800원입니다")
````

````python
#3. if문의 흐름(작동원리)
#코드는 위에 아래 순서대로 작동합니다.

# 2)개수라는 변수에 담음 1)입력을 받음
개수 = int(input("붕어빵 몇개 사실거에요?"))
# 3)개수 변수가 있네 값이 뭐지 4>3  True 또는 False
if 개수>3:
    print("개당 1800원입니다") # 4) 조건이 True이면 if문 안에 코드가 실행됨
if 개수<=3:
    print("개당 2000원입니다")
````

````python
# 4. if 중첩문

의문의_숫자 = 10

if 의문의_숫자 >5:
    if 의문의_숫자<15:
        print("의문의 숫자는 5보다 크고 15보다 작습니다")
````

````python
#5. if ~else 문법과 주의사항(콜론, 들여쓰기)
#True일때는 if문 안에 작성된 코드가 실행됩니다
#False일때는 else문 안에 작성된 코드가 실행됩니다.
개수=int(input("붕어빵 몇개 사실거에요?"))

if 개수>3:
    print("개당 1800원입니다")
else:
    print("개당 2000원입니다")
````

````python
# 7.조건문 조건(True, False)
#True
if True:
    print(True)
else:
    print(False)

#False
if False:
    print(True)
else:
    print(False)

#None
if None:
    print(True)
else:
    print(False)

#0
if 0:
    print(True)
else:
    print(False)

#1
if 1:
    print(True)
else:
    print(False)

#'hello'
if 'hello':
    print(True)
else:
    print(False)

#""
if '':
    print(True)
else:
    print(False)
# not
if not True:
    print(True)
else:
    print(False)

if not '':
    print(True)
else:
    print(False)
````

````python
#9. elif ~ if문
맛 = input("붕어빵 맛은 골라주세요(팥,슈크림,잡채)")

if 맛 == "팥":
    print(f'{맛}을 선택하셨습니다')
elif 맛 == "슈크림":
    print(f'{맛}을 선택하셨습니다')
elif 맛 == "잡채":
    print(f'{맛}을 선택하셨습니다')
else:
    print("다시 골라주세요")
````

````python
# 10. 예제로 if문 익혀보기
#문제 : 단어 a의 가운데 글자를 반환하는 코드
#조건 : 단어의 길이가 짝수라면 가운데 두글자를 출력하면 됩니다.

#입력값1 : "abcde" => c
#입력값2 : "qwer" => we

a = input()
# 코드를 작성해주세요

가운데_인덱스 = len(a)//2

if len(a)%2!=0:
    print(a[가운데_인덱스])
else:
    print(a[가운데_인덱스-1 : 가운데_인덱스+1])
````

```python
#예외코드 작성

try:
    x=int(input('나눌 숫자를 입력하세요'))
    y=10/x
    print(y)
except:
    print('예외가 발생했습니다.')
```

```python
y=[10,20,30]
try:
    index,x = map(int, input('나눌 숫자를 입력하세요: ').split(','))
    print(y[index]/x)
except ZeroDivisionError as e:
    print('숫자를 0으로 나눌 수 없습니다.',e)
except IndexError as e:
    print('잘못된 인덱스입니다.',e)
#
```

```python
#예외코드 작성

try:
    x=int(input('나눌 숫자를 입력하세요'))
    y=10/x
    raise
    print(y)
except:
    print('예외가 발생했습니다.')
else:
    print("성공하셨습니다")
finally:
    print('예외의 여부와 상관없이 항상 실행됩니다.')
#
```

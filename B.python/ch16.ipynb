```python
#1. 만들어보기

i=0
while i<10:
    print('점점 성장하는 여러분',i)
    i += 1
#
```

```python
i=0
while i<10:
    print("동물")
    print("보러 갈래요?")
    i+=1
```

```python
# 3. while문 조건식 증강,감소

# 1. 기본적인 증가 유형

i=1
while i<10:
    print("이정도로 성장할 줄이야",i)
    i+=1

#2. 기본적인 감소 유형
i=10
while i>0:
    print("이정도로 성장할 줄이야",i)
    i-=1
```

```python
#4. 조심스럽게 사용해야 하는 이유
# 조건이 잘못 설정되면 무한 루프에 빠져 과도한 리소스 소모
# 하지만 잘 사용하면 for문보다 효율적이다

# while True:
    # print("무한루프",1)
    # break

while True:
    print("돈복사","내가 산 주식 폭등")
    break
```

```python
# 5. while 문이 for문보다 강점인 부분?
# 무수히 많이 실행해야 할때

import random
# random.randint(1,6)
i=0
while i!=4:
    i =random.randint(1,6)
    print(i)
```

```python
# 6 while 문에도 else가 있어요

양파=10
while 양파>10:
    print("양파", 양파, "개 남았습니다.")
    if 양파==0:
        break
    양파-=1
else:
    print("양파가 다 떨어졌습니다.")
    print("판매를 중지합니다.")
```

```python
#7. 반복문에 흐름을 제어하는 continue와 break

i=0
while True:
    print(i)
    i += 1

    if i==10:
        break

for i in range(100):
    print(i)
    if i ==10:
        break
```

```python
i=0
while i<10:
    i+=1
    if i%2==0:
        continue
    print(i)

for i in range(10):
    if i%2==0:
        continue
    print(i)
```

```python
# 9. 예제로 while문 익혀보기

#콜라츠하는 사람은 주어진 수가 1이 될때까지 다음 작업을 반복하면, 모든 수를 1로 만들 수 있다고 추측했다
#작업
#1-1. 입력된 수가 짝수라면 2로 나눈다.
#1-2. 입력된 수가 홀수라면 3을 곱하고 1을 더한다.
#1-3. 결과로 나온 수에 같은 작업을 1이 될때까지 반복한다.
#예를 들어, 입력된 수가 6이라면 6→3→10→5→16→8→4→2→1 이 됩니다.

# num 변수에 있는 수를 1로 만드는데 몇번이나 작업을 반복해야 하는지 출력한다
num=7

정답 = 0
while num!=1:
    if num%2==0:
        num=num//2
    else:
        num=num*3+1
    정답+=1
print(정답)
```

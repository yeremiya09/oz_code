```python
# (기본 매개변수, 가변 매개변수)
def oz(n=3, *values):
    for i in range(n):
        for value in values:
            print(value)

oz("기본 매개변수 다음에", "가변 매개변수를", "넣을수 있을까요?")
```

```python
# (가변 매개변수, 기본 매개변수)
def oz(*values, n=3):
    for i in range(n):
        for value in values:
            print(value)

oz("기본 매개변수 다음에", "가변 매개변수를", "넣을수 있을까요?", 10)
```

```python
# 키워드 매개변수
list = ["안", "녕", "하", "세", "요"]
for i in list:
    print(i)
```

```python
# 키워드 매개변수
list = ["안", "녕", "하", "세", "요"]
for i in list:
    print(i, end="")
```

```python
#  직육면체를 부피를 구하는 코드를 작성해보자: 가로 : x, 세로 : y, 높이 : h
def oz(x, y=20, h=10):
    print(x*y*h)

print(oz(10))

print(oz(10, y=10))

print(oz(x=5, y=5, h=5))
# print 문을 지우면 none이 없어진다
```

```python
import datetime

print("오늘 날짜와 시간 출력")
now = datetime.datetime.now()
print(now.year, "년")
print(now.month, "월")
print(now.day, "일")
print(now.hour, "시")
print(now.minute, "분")
print(now.second, "초")

print(now.strftime("%Y년 %m월 %d일 %H시 %M분 %S초"))
```

```python
import datetime

print("다양한 시간 포멧")
output_1 = now.strftime("%Y.%m.%d %H:%M:%S")
output_2 = f'{now.year}년.{now.month}월.{now.day}일 {now.hour}시:{now.minute}분:{now.second}초'
output_3 = now.strftime("%Y{} %m{} %d{} %H{} %M{} %S{}").format(*"년월일시분초")

print(output_1)
print(output_2)
print(output_3)
print()
```

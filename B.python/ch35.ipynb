```python
try:
    x, y = map(int, input("삼각형의 밑변과 높이를 입력하세요 : ").split())
    print(f'삼각형의 넓이는 {x*y/2} 입니다')
except Exception as exception:
    print("type(excerption):", type(exception))
    print("exception:", exception)
```

```python
numbers = [23, 11, 7, 4, 12]

try:
    number_input = int(input("찾고싶은 값의 위치를 입력해주세요"))
    print(f'{number_input}의 위치는 {numbers.index(number_input)} 입니다')
except Exception as exception:
    print("type(excerption):", type(exception))
    print("exception:", exception)
```

```python
numbers = [23, 11, 7, 4, 12]

try:
    number_input = int(input("찾고싶은 값의 위치를 입력해주세요"))
    print(f'{number_input}번쨔 요소: {numbers[number_input]}')
except ValueError:
    print("정수로 입력해주세요")
except IndexError:
    print("리스트의 범위를 벗어났습니다. 입력값을 다시 확인해주세요")
```

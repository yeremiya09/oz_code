##시퀀스 자료형
* 문자열 :  
        * 시퀀스 형에서 사용할 수 있는 메소드(count(), index(), split() 등...)
        * 슬라이싱([:],[::])
        * 요소들로 이루어져 가능했던 기능들(요소 합치기, 요소 찾기 등등)
* list :  
        * 시퀀스 형에서 사용할 수 있는 메소드(count(), index(), split() 등...)
        * 슬라이싱([:],[::])
        * 요소들로 이루어져 가능했던 기능들(요소 합치기, 요소 찾기 등등)
        * list 지원하는 다양한 메소드(append(), insert(), del, sort(), reverse())
* tuple :
        * 리스트처럼 요소를 일렬로 저장하지만 안에 저장된 요소를 변경, 추가, 삭제할 수 없다
        * 값을 추가, 삭제, 변경할 수 없기 때문에, 사용하지 못하는 메소드가 많다
        * tuple의 생성 방법은 tuple = 값, 값 또는 tuple = (값, 값)
        * 그리고 엄청난 비밀이 숨겨져 있어요
        * 튜플은 소괄호입니다
* dict(딕셔너리)
        * list,tuple,str,range(범위 안에 연속된 데이터를 만드는 시퀀스였습니다)
        공통적인 특징이 있어요 데이터의 연관성이 없습니다
        * dict는 여러개의 값이 일렬로 정렬되면서도 값끼리의 연관성이 존재
        (예시 : 게임 캐릭터 힘:30, 지능:20, 체력:40, 민첩:90 )
        * dict 생성방법은 중괄호

```python
#list[], tuple(), dict{}

짱짱맨 ={"힘":30, "지능":20, "체력":40, "민첩":90}
짱짱맨

#{키 : 값}
#{key1:value1, key2:value2}

#시퀀스 데이터는 인덱싱을 이용해 데이터를 찾았습니다 dict는 다른 방법
#짱짱맨[1]
짱짱맨["힘"]
print(f'짱짱맨님은 힘이 {짱짱맨["힘"]}입니다.')
```

```python
#dict 이력서를 작성했어요. 붕어빵인터네셔널 채용공고 열정만 있으면 오세요
# 바로 오라고 하는게 이상해요
# 아무도 없이 코드만 있어요... 레거시
# 리팩토링 해달라고 합니다.

# 변수값 입력
fish_p, fish_s, fish_j = "붕어빵(팥)", "붕어빵(슈크림)", "붕어빵(잡채)"
fish_p_sell, fish_s_sell, fish_j_sell = 2000, 2500, 3000
order_p, order_s, order_j = 0, 2, 3
# 계산값
order_p_pay, order_s_pay, order_j_pay, = fish_p_sell*order_p, fish_s_sell*order_s, fish_j_sell*order_j
total = order_p + order_s + order_j
total_pay=order_p_pay + order_s_pay + order_j_pay

print(f'주문이 완료되었습니다 {fish_s} {order_s}개 {fish_j} {order_j}개 총 {total}개 걸제 금액은 {total_pay}원 입니다')
```

```python
주문_개수=map(int,input(f'붕어빵 맛 순서대로 개수를 입력해주세요 안 살거면 0입력(팥,슈크림,잡채)').split(','))

품목별_가격 = {"팥":2000, "슈크림":2500, "잡채":3000}
#딕셔너리는 순서가 없어 인덱싱으로 접근 불가

주문_개수 = list(주문_개수)
총_주문_개수 = sum(주문_개수)
total=0

total += 품목별_가격["팥"]*주문_개수[0]
total += 품목별_가격["슈크림"]*주문_개수[1]
total += 품목별_가격["잡채"]*주문_개수[2]

print(f'주문이 완료되었습니다. 팥{주문_개수[0]}개, 슈크림{주문_개수[1]}개, 잡채{주문_개수[2]}개 총{총_주문_개수}개 결제금액은{total:,}원 입니다 ')
```
